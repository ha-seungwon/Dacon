## Import

```python
import pandas as pd
import random
import os
import numpy as np
from xgboost import XGBClassifier
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
class CFG:
    SEED = 42
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(CFG.SEED) # Seed 고정
```

## Data Load

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
train=train.drop(['father','mother','gender'],axis=1)
test=test.drop(['father','mother','gender'],axis=1)
```

```python
def get_x_y(df):
    if 'class' in df.columns:
        df_x = df.drop(columns=['id', 'class'])
        df_y = df['class']
        return df_x, df_y
    else:
        df_x = df.drop(columns=['id'])
        return df_x
```

```python
train_x, train_y = get_x_y(train)
test_x = get_x_y(test)
```

## Data Pre-processing
### Label-Encoding

```python
class_le = preprocessing.LabelEncoder()
snp_le = preprocessing.LabelEncoder()
snp_col = [f'SNP_{str(x).zfill(2)}' for x in range(1,16)]
```

```python
snp_data = []
for col in snp_col:
    snp_data += list(train_x[col].values)
```

```python
train_y = class_le.fit_transform(train_y)
snp_le.fit(snp_data)
```

```python
for col in train_x.columns:
    if col in snp_col:
        train_x[col] = snp_le.transform(train_x[col])
        test_x[col] = snp_le.transform(test_x[col])
```

```python

train_x=pd.get_dummies(train_x,columns=train_x.columns)

test_x=pd.get_dummies(test_x,columns=test_x.columns)
```

```python
train_x
```

```python
test_x
```

## Model Fit

```python
clf = XGBClassifier(random_state=CFG.SEED)
clf.fit(train_x, train_y)
```

## Inference

```python
preds = clf.predict(test_x)
print(preds)
```

## Submission

```python
submit = pd.read_csv('./sample_submission.csv')
```

```python
submit['class'] = class_le.inverse_transform(preds)
```

```python
submit.to_csv('./submit.csv', index=False)
```
